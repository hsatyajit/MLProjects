```python
# Transform the data
# Transform non-numeric column into numeric column
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

```python
df
```

```python
df['Age_Years'] = df['Age']
```

```python
df = df.drop('Age', axis=1)
```

```python
df.head()
```

```python
#Split the data
X = df.iloc[:,1:df.shape[1]].values
y = df.iloc[:,0].values
```

```python
# Split the into 75% training and 25% testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state= 0)
```

```python
from pandas.core.common import random_state
# Use the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators= 10, criterion= 'entropy', random_state= 0)
forest.fit(X_train, y_train)
```

```python
# Get the accuracy of the training dataset
forest.score(X_train, y_train)
```

```python
# Show the accuracy and the confusin matrix for the model on the test dataset
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

disp = plot_confusion_matrix(forest, X_test, y_test, cmap='Blues')
cm = confusion_matrix(y_test, forest.predict(X_test))

TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]

print(cm)
print('Model accuracy = {}'.format((TP+TN)/(TP+TN+FN+FP)))
```
